8.	Ejecuta el siguiente código y comenta qué hace cada línea:

Cada linea hace lo siguiente:

1. `datetime_df = air_df.ReadingDateTime.str.split(' ', expand=True)`  
   Esta línea toma la columna `ReadingDateTime` y divide cada entrada en dos partes separadas por un espacio: la fecha y la hora. El argumento `expand=True` hace que el resultado sea un nuevo DataFrame donde cada parte (fecha y hora) es una columna separada.

2. `datetime_df.columns = ['Date','Time']`  
   Asigna nombres a las dos columnas resultantes del split anterior, llamándolas `Date` y `Time`.

3. `datetime_df`  
   Muestra el DataFrame

4. `date_df = datetime_df.Date.str.split('/', expand=True)`  
   Toma la columna `Date` de `datetime_df`, que contiene las fechas, y las divide por el carácter `/`. Esto separa la fecha en día, mes y año, poniendo cada parte en su propia columna en un nuevo DataFrame llamado `date_df`.

5. `date_df.columns = ['Day','Month','Year']`  
   Nombra las columnas del DataFrame `date_df` como `Day`, `Month` y `Year`, correspondiendo a las partes de la fecha divididas en la línea anterior.

6. `date_df`  
   Muestra el DataFrame `date_df`.

7. `air_df = air_df.join(date_df).join(datetime_df.Time).drop(columns=['ReadingDateTime','Year'])`  
   Esta línea hace varias instrucciones:
   Junta (`join`) el `air_df` original con `date_df` para añadir las columnas `Day`, `Month` y `Year`.
   Realiza otra unión para añadir la columna `Time` de `datetime_df` a `air_df`.
   Con `drop`, elimina las columnas `ReadingDateTime` y `Year` del DataFrame resultante porque ya no son necesarias.

8. `air_df = air_df.set_index(['Month','Day','Time','Species'])`  
   Establece un índice multinivel con las columnas `Month`, `Day`, `Time` y `Species`, lo que permite un acceso rápido y eficiente a los datos basado en estos índices jerárquicos.

En resumen todo el código sirve para crear un df donde las mediciones están organizadas y accesibles por mes, día, hora y contaminante


```python
datetime_df = air_df.ReadingDateTime.str.split(' ',expand=True)
datetime_df.columns = ['Date','Time']
datetime_df
date_df = datetime_df.Date.str.split('/',expand=True)
date_df.columns = ['Day','Month','Year']
date_df
air_df = air_df.join(date_df).join(datetime_df.Time).drop(columns=['ReadingDateTime','Year'])
air_df = air_df.set_index(['Month','Day','Time','Species'])
air_df
```

10.	Explica cuáles son las diferencias y similitudes entre el par `melt()/pivot()` y el par `stack()/unstack()`. Si tuvieras que elegir una contraparte para `melt()` entre `stack()/unstack()`, ¿cuál elegirías? ¿Por qué?

Melt y pivot son funciones que se utilizan para transformar df entre formatos "largo" y "ancho". Stack y unstack hacen algo similar pero trabajan a nivel de índices, operando dentro de los niveles con índice multinivel.

Similitudes:
Tanto melt como stack se utilizan para transformar datos de un formato ancho a un formato largo.
Tanto pivot como unstack hacen lo contrario, transformando datos de un formato largo a uno ancho.

Diferencias:
Melt es una función que toma un DataFrame y lo transforma en un formato donde una o más columnas son identificadores
Pivot toma un DataFrame en formato largo y lo extiende a un formato ancho
Stack convierte columnas en índices, apilando los datos y transformando las columnas de nivel más bajo en el índice más interno.
Unstack toma un nivel de índice y lo pivota,

Elegiría stack sobre melt para trabajar con índices multinivel y mantener una relación jerárquica entre las columnas convertidas y los índices.
